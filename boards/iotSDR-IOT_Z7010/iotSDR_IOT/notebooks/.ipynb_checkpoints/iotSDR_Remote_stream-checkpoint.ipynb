```python
import iotSDR_streaming as iotSDR_stream
import iotSDR_defs as defs
from pynq import Overlay
import matplotlib.pyplot as plt

ol = Overlay('/home/xilinx/pynq/overlays/base/dualRxTxDmaOTAv1.3.1.bit', download=True)

iotSDR = iotSDR_stream.iotSDR_stream(ol,0)
#status = iotSDR.connect("192.168.2.99:42400")

rx_chan   = defs.chan_subGHz_B
tx_chan   = defs.chan_subGHz_A
```

```python
import numpy as np
import time
# Update TX LUT
f = 5 #Hz
N = 1024

t = np.linspace(0, 1, N,endpoint=False)
real = np.cos(2*np.pi*f*t)
imag = np.sin(2*np.pi*f*t)

IQsamples = real + 1j*imag

plt.plot(t, real[:])
plt.show()

#ol = Overlay('/home/xilinx/pynq/overlays/base/dualRxTxDmaOTAv1.3.1.bit', download=True)

iotSDR.setTxLUT(tx_chan,IQsamples)
s#chA = iotFrontEndCore.Transceiver("ch1",ol)
#chA.set_tx_LUT(IQsamples)
```

```python
freq = 412e6
iotSDR.setFrequency(tx_chan,freq)
iotSDR.setFrequency(rx_chan,freq)
#print(iotSDR.getFrequency(tx_chan))

rx_streamA = iotSDR.setupStream(defs.IOTSDR_RX,rx_chan)
```
