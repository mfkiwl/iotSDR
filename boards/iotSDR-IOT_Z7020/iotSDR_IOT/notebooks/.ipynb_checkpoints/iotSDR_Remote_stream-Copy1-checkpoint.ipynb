```python
from iotSDR_Remote import * #iotSDR_Remote as remote
import iotSDR_defs as defs
import matplotlib.pyplot as plt

iotSDR = iotSDR_Remote()
status = iotSDR.connect("localhost:42400")

rx_chan   = defs.chan_subGHz_B
tx_chan   = defs.chan_subGHz_A
```

```python
import numpy as np
import time
# Update TX LUT
f = 5 #Hz
N = 1024

t = np.linspace(0, 1, N,endpoint=False)
real = np.cos(2*np.pi*f*t)
imag = np.sin(2*np.pi*f*t)

IQsamples = real + 1j*imag

plt.plot(t, real[:])
plt.show()

iotSDR.setTxLUT(tx_chan,IQsamples)
```

```python
freq = 412e6
iotSDR.setFrequency(tx_chan,freq)
iotSDR.setFrequency(rx_chan,freq)
#print(iotSDR.getFrequency(tx_chan))

rx_streamA = iotSDR.setupStream(defs.IOTSDR_RX,rx_chan)
print(rx_streamA)
```

```python
import time

count = 0
totallen = 0
samplesPerFrame = 2*4*1024
numOfFrames = 2

tic = time.clock()
for frame in (iotSDR.readStream(rx_chan,numOfFrames)):
    totallen += len(frame.frame)
    frame = np.frombuffer(frame.frame)
    #print(frame,len(frame))
    count +=1

toc = time.clock()
print(count,totallen/samplesPerFrame,toc-tic)
```
